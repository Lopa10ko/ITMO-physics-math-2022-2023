‚ñ∂ –ù–∞—á–Ω–µ–º —Å **–∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–π —Ç–µ–æ—Ä–∏–∏ —Ç—è–≥–æ—Ç–µ–Ω–∏—è**:
---
*–ù–∞–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç—å –≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª—è* - —ç—Ç–æ –≤–µ–∫—Ç–æ—Ä–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞, —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É—é—â–∞—è —ç—Ç–æ –ø–æ–ª–µ –≤ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Ç–æ—á–∫–µ –∏ —á–∏—Å–ª–µ–Ω–Ω–æ —Ä–∞–≤–Ω–∞—è –æ—Ç–Ω–æ—à–µ–Ω–∏—é –≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–π —Å–∏–ª—ã $F$, –¥–µ–π—Å—Ç–≤—É—é—â–µ–π –Ω–∞ –Ω–µ–ø–æ–¥–≤–∏–∂–Ω—É—é –ø—Ä–æ–±–Ω—É—é —á–∞—Å—Ç–∏—Ü—É —ç—Ç–∞–ª–æ–Ω–Ω–æ–π –º–∞—Å—Å—ã $m_0$ –≤ —ç—Ç–æ–π —Ç–æ—á–∫–µ. \
–¢–æ–≥–¥–∞ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è –Ω–∞–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏ –≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª—è –≤—ã–≥–ª—è–¥–∏—Ç:
\begin{equation}
  \Gamma=\frac{F}{m_0} 
\end{equation}
–í –∫–ª–∞—Å—Å–∏—á–µ—Å–∫–æ–π —Ç–µ–æ—Ä–∏–∏ —Ç—è–≥–æ—Ç–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–µ –≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–π —Å–∏–ª—ã –º–æ–∂–µ—Ç –±—ã—Ç—å –∑–∞–ø–∏—Å–∞–Ω–æ –≤ —É–ø—Ä–æ—â–µ–Ω–Ω–æ–π —Ñ–æ—Ä–º–µ, –µ—Å–ª–∏ –ø–æ–ª–∞–≥–∞—Ç—å, —á—Ç–æ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–º –≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª—è —è–≤–ª—è–µ—Ç—Å—è –æ–¥–Ω–æ—Ä–æ–¥–Ω–æ–µ —Ç–µ–ª–æ —Å—Ñ–µ—Ä–∏—á–µ—Å–∫–æ–π —Ñ–æ—Ä–º—ã –º–∞—Å—Å–æ–π $M_–ó$ –∏ —Ä–∞–¥–∏—É—Å–æ–º $R_–ó$, —Ç–æ –µ—Å—Ç—å:
\begin{equation}
  \Gamma=\frac{-\frac{Gm_0M_–ó}{\left(R_–ó+r\right)^2}}{m_0} = -\frac{GM_–ó}{\left(R_–ó+r\right)^2} , \: \: \: \: r \in \mathbb{R_+ \cup \{0\}}
\end{equation}
–ó–∞–º–µ—Ç–∏–º, —á—Ç–æ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å $\Gamma\left(R\right)$ –≤–∞–ª–∏–¥–Ω–∞ –¥–ª—è —Ç–æ—á–µ–∫ –∑–∞ –≥—Ä–∞–Ω–∏—Ü–µ–π –º–æ–¥–µ–ª–∏—Ä—É–µ–º–æ–π —Å—Ñ–µ—Ä—ã –ó–µ–º–ª–∏. \
–ù–∞–ø—Ä–∏–º–µ—Ä, –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª —Ç–æ—á–∫–∏ –∑–∞ —Å—Ñ–µ—Ä–æ–π —Ä–∞–≤–µ–Ω $\psi = -\frac{GM_–ó}{R_–ó+r}$, —Ç–æ–≥–¥–∞:
\begin{equation}
  \overrightarrow{\Gamma}\xrightarrow[-\nabla\psi]{}-\frac{GM_–ó}{\left(R_–ó+r\right)^3} \cdot \overrightarrow{\left(R_–ó+r\right)}
\end{equation}
–ò–ª–∏, —á—Ç–æ —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ —á–µ—Ä–µ–∑ –ø—Ä–∏–Ω—Ü–∏–ø —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–Ω–æ—Å—Ç–∏ –∏–Ω–µ—Ä—Ü–∏–∞–ª—å–Ω–æ–π –∏ –≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Å—Å:
\begin{equation}
  \overrightarrow{F} = m_0\overrightarrow{g} = -\frac{Gm_0M_–ó}{\left(R_–ó+r\right)^3} \cdot \overrightarrow{\left(R_–ó+r\right)} \Rightarrow \overrightarrow{g} \equiv \overrightarrow{\Gamma} =-\frac{GM_–ó}{\left(R_–ó+r\right)^3} \cdot \overrightarrow{\left(R_–ó+r\right)}
\end{equation}
–ó–∞–º–µ—Ç–∏–º, —á—Ç–æ –≤–Ω—É—Ç—Ä–∏ —Å—Ñ–µ—Ä—ã (–≥–µ–æ–∏–¥–∞ –ó–µ–º–ª–∏) –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª –ø–æ—Å—Ç–æ—è–Ω–µ–Ω, —Ç–æ –µ—Å—Ç—å –Ω–∞–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç—å –ª–∏–Ω–µ–π–Ω–æ —Ä–∞—Å—Ç–µ—Ç –¥–æ –∑–Ω–∞—á–µ–Ω–∏—è $\left[-\frac{GM_–ó}{R_–ó^2}\right]$. \
–¢–æ –µ—Å—Ç—å –æ–ø–∏—Å–∞–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –Ω–∞–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏ –≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª—è –æ—Ç —Ä–∞–¥–∏—É—Å-–≤–µ–∫—Ç–æ—Ä–∞ –¥–æ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Ç–æ—á–∫–∏ –º–æ–∂–Ω–æ –¥–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–º –æ–±—Ä–∞–∑–æ–º:
\begin{equation}
  \Gamma =\left\{\begin{matrix}
-\frac{GM_–ó}{R_–ó^3} \cdot r, \: \: \: \: r \in \left[0; R_–ó \right]\\ 
-\frac{GM_–ó}{r^2}, \: \: \: \:   r > R_–ó
\end{matrix}\right.
\end{equation}

–¢–µ–ø–µ—Ä—å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª–∞ –≥—Ä–∞–≤–∏—Ç–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª—è, –∫–æ—Ç–æ—Ä–æ–µ —Å–æ–∑–¥–∞–µ—Ç—Å—è –Ω–∞—à–∏–º –æ–¥–Ω–æ—Ä–æ–¥–Ω—ã–º —à–∞—Ä–æ–º - **–ó–µ–º–ª–µ–π**: \
*   –°–Ω–∞—Ä—É–∂–∏ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª —Ä–∞–≤–µ–Ω $\psi = -\frac{GM_–ó}{r}, \: \: r \in \left(R_–ó, +\infty \right)$
*   –í–Ω—É—Ç—Ä–∏ —à–∞—Ä–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞—Ç—å —à–∞—Ä–æ–≤–æ–π —Å–ª–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π –º–∞—Å—Å—ã, —Ç–æ–≥–¥–∞ –¥–ª—è —ç—Ç–æ–≥–æ —Å–ª—É—á–∞—è –≤—ã—Ä–∞–∂–µ–Ω–∏–µ —Å–∏–ª—ã —Ç—è–≥–æ—Ç–µ–Ω–∏—è –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è $|F| = \frac{Gm_0M_{layer}}{r^2} = Gm_0\cdot \left(\frac{4\rho\pi}{3} \right) \cdot r$, –∞ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª–∞: $\psi = G \cdot \frac{2\pi \rho}{3} \cdot r^2 - 2\pi \rho R_–ó^2, \: \: r \in \left[0,R_–ó \right]$
\begin{equation}
  \psi =\left\{\begin{matrix}
\frac{GM_–ór^2}{2R_–ó^3} - \frac{3GM_–ó}{2R_–ó}, \: \: \: \: r \in \left[0; R_–ó \right]\\ 
-\frac{GM_–ó}{r}, \: \: \: \:   r > R_–ó
\end{matrix}\right.
\end{equation}
–ó–∞–º–µ—Ç–∏–º –µ—â–µ, —á—Ç–æ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–∞—è —ç–Ω–µ—Ä–≥–∏—è –¥–ª—è –µ–¥–∏–Ω–∏—á–Ω–æ–π –º–∞—Å—Å—ã –±—É–¥–µ—Ç –ø–æ –≥—Ä–∞—Ñ–∏–∫—É —Å–æ–ø–∞–¥–∞—Ç—å —Å –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª–æ–º —Å —Ç–æ—á–Ω–æ—Å—Ç—å—é –¥–æ —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç–µ–π, –∏–∑ —Ä–∞–∑–Ω–æ—Å—Ç–µ–π –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã—Ö —ç–Ω–µ—Ä–≥–∏–π –ø–æ–Ω—è—Ç–Ω–æ, —á—Ç–æ –Ω–∞ –ø–æ–≤–µ—Ä—Ö–Ω–æ—Å—Ç–∏ –ó–µ–º–ª–∏ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω–∞—è —ç–Ω–µ—Ä–≥–∏—è —Ä–∞–≤–Ω–∞ —Ä–∞–±–æ—Ç–µ —Å–∏–ª—ã —Ç—è–∂–µ—Å—Ç–∏ $mgh$.

```python
class Vector(list):
    def __init__(self, *el):
        for e in el:
            self.append(e)
    def __add__(self, other):
        if type(other) is Vector:
            assert len(self) == len(other), "Error 0"
            r = Vector()
            for i in range(len(self)):
                r.append(self[i] + other[i])
            return r
        else:
            other = Vector.emptyvec(lens=len(self), n=other)
            return self + other
    def __sub__(self, other):
        if type(other) is Vector:
            assert len(self) == len(other), "Error 0"
            r = Vector()
            for i in range(len(self)):
                r.append(self[i] - other[i])
            return r
        else:
            other = Vector.emptyvec(lens=len(self), n=other)
            return self - other
    def __mul__(self, other):
        if type(other) is Vector:
            assert len(self) == len(other), "Error 0"
            r = Vector()
            for i in range(len(self)):
                r.append(self[i] * other[i])
            return r
        else:
            other = Vector.emptyvec(lens=len(self), n=other)
            return self * other
    def __truediv__(self, other):
        if type(other) is Vector:
            assert len(self) == len(other), "Error 0"
            r = Vector()
            for i in range(len(self)):
                r.append(self[i] / other[i])
            return r
        else:
            other = Vector.emptyvec(lens=len(self), n=other)
            return self / other
    def __pow__(self, other):
        if type(other) is Vector:
            assert len(self) == len(other), "Error 0"
            r = Vector()
            for i in range(len(self)):
                r.append(self[i] ** other[i])
            return r
        else:
            other = Vector.emptyvec(lens=len(self), n=other)
            return self ** other
    def __mod__(self, other):
        return sum((self - other) ** 2) ** 0.5
    def mod(self):
        return self % Vector.emptyvec(len(self))
    def dim(self):
        return len(self)
    def __str__(self):
        if len(self) == 0:
            return "Empty"
        r = [str(i) for i in self]
        return "< " + " ".join(r) + " >" 
    def _ipython_display_(self):
        print(str(self))
    @staticmethod
    def emptyvec(lens=2, n=0):
        return Vector(*[n for i in range(lens)])
    @staticmethod
    def randvec(dim):
        return Vector(*[random.random() for i in range(dim)])
    @staticmethod
    def centervec(dim):
        return Vector(*[0.5 for i in range(dim)])
```

```python
class Point:
    def __init__(self, coords, mass=1.0, q=1.0, speed=None, **properties):
        self.coords = coords
        if speed is None:
            self.speed = Vector(*[0 for i in range(len(coords))])
        else:
            self.speed = speed
        self.acc = Vector(*[0 for i in range(len(coords))])
        self.mass = mass
        self.__params__ = ["coords", "speed", "acc", "q"] + list(properties.keys())
        self.q = q
        for prop in properties:
            setattr(self, prop, properties[prop])
    def move(self, dt):
        self.coords = self.coords + self.speed * dt
    def accelerate(self, dt):
        self.speed = self.speed + self.acc * dt
    def accinc(self, force):
        self.acc = self.acc + force / self.mass
    def clean_acc(self):
        self.acc = self.acc * 0
    def __str__(self):
        r = ["Point {"]
        for p in self.__params__:
            r.append("  " + p + " = " + str(getattr(self, p)))
        r += ["}"]
        return "\n".join(r)
    def _ipython_display_(self):
        print(str(self))
```

```python
class InteractionField:
    def __init__(self, F):
        self.points = []
        self.F = F
    def move_all(self, dt):
        for p in self.points:
            p.move(dt)
    def intensity(self, coord):
        proj = Vector(*[0 for i in range(coord.dim())])
        single_point = Point(Vector(), mass=1.0, q=1.0)
        for p in self.points:
            if coord % p.coords < 10 ** (-10):
                continue
            d = p.coords % coord
            fmod = self.F(single_point, p, d) * (-1)
            proj = proj + (coord - p.coords) / d * fmod
        return proj
    def step(self, dt):
        self.clean_acc()
        for p in self.points:
            p.accinc(self.intensity(p.coords) * p.q)
            p.accelerate(dt)
            p.move(dt)
    def clean_acc(self):
        for p in self.points:
            p.clean_acc()
    def append(self, *args, **kwargs):
        self.points.append(Point(*args, **kwargs))
    def gather_coords(self):
        return [p.coords for p in self.points]
```
